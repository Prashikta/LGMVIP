## Task-3: Prediction using Decision Tree  Algorithm :

Create the Decision Tree classifier and visualize it graphically. 
The purpose is if we feed any new data to this classifier, it would be able to  predict the right class accordingly.



### Importing required libraries

```python
#Importing the required libraries
import pandas as pd
import seaborn as sns
```

### Reading the dataset

```python
# Loading the iris dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
data = pd.read_csv(url, names=names)
```

```python
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#Frequency Table
data["class"].value_counts()
```

This tell us that there are 3 classes.

### Data Visualization

```python
#ploting pairplot
sns.pairplot(data = data, hue='class')
```

```python
data.corr()
```

```python
#plotting Heatmap
sns.heatmap(data.corr(),annot= True)
```

```python
#Dividing the dataset into dependent and independent variables
x = data.iloc[:,:4].values
y = data.iloc[:,4:5]
```

```python
#Spliting the data into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

###Model Building

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Accuracy of the model
print("Accuracy: ", metrics.accuracy_score(Y_test,y_pred))
```

This means our model is 100% accurate.

####Predicting the right class on the new data feeded.

```python
#predicting class for new data
new_data = [[4.6, 2.0, 4.2, 1.1]]
y_pred = model.predict(new_data)
print(y_pred)
```

###Visualising decision tree

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize = (20,10))
tree.plot_tree(model, filled=True, rounded=True)
plt.show()
```
